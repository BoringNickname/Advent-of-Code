```python
import re

with open('input.txt') as file:
    lines = [line.rstrip() for line in file]
    
lines = [re.sub(r'^.*?: ', '', line) for line in lines]
```

```python
point_sum = 0
for i, line in enumerate(lines):
    split_line = line.split('|')
    #now we split the line into winning and personal numbers and map each str to an int for each number string
    winning_numbers = list(map(int, re.findall(r'\d+', split_line[0])))
    my_numbers = list(map(int, re.findall(r'\d+', split_line[1])))
    
    matches = 0
    for number in my_numbers: #check if it contains a winning number
        if number in winning_numbers:
            matches+=1
    if matches != 0: #formula only works if we have more than 0 matches 
        point_sum+=2**(matches-1)
    print(f'line {i+1} has {matches} matches')
print(point_sum)
```
